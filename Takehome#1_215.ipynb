## Plot each data point on a Catesian plane

```python
import matplotlib.pyplot as plt

# Data from the table
data = {
    'May 6th, 2024': [(0, 51), (12, 70), (20, 81), (28, 90)],
    'May 8th, 2024': [(0, 32), (7, 50), (15, 66), (20, 75), (30, 88), (32, 90)],
    'May 13th, 2024': [(0, 12), (6, 33), (13, 52), (18, 62), (29, 78), (35, 85), (40, 90)]
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot the data points for each date
for date, points in data.items():
    times = [point[0] for point in points]
    percentages = [point[1] for point in points]
    plt.plot(times, percentages, marker='o', label=date)

# Adding titles and labels
plt.title('Tesla Model Y Battery Charge Data')
plt.xlabel('Time (minutes)')
plt.ylabel('Battery Percentage (%)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

There is a positive correlation between time charging and battery percentage (as one increases the other increases).  Though the correlation is positive the shape does not indicate a linear relationship, so a non-linear relationship would be best to describe the relationship between time and battery percentage.  Another observation is that as time increases the battery percentage does not seem to increase as fast as when it started charging.  

Your answers will vary when selecting two points for creating the linear functiono, but in my opinion since there is a starting and ending, an average might be a good place to start.  So the point would be (0,31.7) and (33.3, 90).  Thus the linear function is

$$B(t) = 1.75t + 31.7$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
data = {
    'May 6th, 2024': [(0, 51), (12, 70), (20, 81), (28, 90)],
    'May 8th, 2024': [(0, 32), (7, 50), (15, 66), (20, 75), (30, 88), (32, 90)],
    'May 13th, 2024': [(0, 12), (6, 33), (13, 52), (18, 62), (29, 78), (35, 85), (40, 90)]
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot the data points for each date
for date, points in data.items():
    times = [point[0] for point in points]
    percentages = [point[1] for point in points]
    plt.plot(times, percentages, marker='o', label=date)

# Add the equation of the line y = 1.75t + 31.7
t = np.linspace(0, 40, 400)  # Generate 400 points between 0 and 40
y = 1.75 * t + 31.7
plt.plot(t, y, label='$y = 1.75t + 31.7$', color='black', linestyle='--')

# Adding titles and labels
plt.title('Tesla Model Y Battery Charge Data with Line Equation')
plt.xlabel('Time (minutes)')
plt.ylabel('Battery Percentage (%)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

$$B(10) = 1.75(10)+31.7 $$
$$B(10) = 49.2$$
Thus I would expect about 49.2\% of the battery to be full after 10 minutes.

$$B(70) =1.75(70)+31.5$$
$$B(70) =154.2$$
Thus I would expect about 154.2\% of the battery to be full after 70 minutes.  Which of course if impossible, so...
